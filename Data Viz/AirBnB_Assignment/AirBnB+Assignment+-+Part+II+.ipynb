### Task 6

You wish to identify the relationship between the various ratings each of the listing has received from the customers.
These ratings have been summarized below:

```python
review_columns = ['review_scores_rating',
 'review_scores_accuracy',
 'review_scores_cleanliness',
 'review_scores_checkin',
 'review_scores_communication',
 'review_scores_location',
 'review_scores_value']
```

Plot a heatmap of the correlation matrix of the above ratings variables and document your observations. You can use this link to understand these variables further - https://www.airbnb.co.in/help/article/1257/star-ratings

```python
corr = df[['review_scores_rating',
 'review_scores_accuracy',
 'review_scores_cleanliness',
 'review_scores_checkin',
 'review_scores_communication',
 'review_scores_location',
 'review_scores_value']].corr()

sns.heatmap(corr,cmap="Blues")
```

**Observation** - 
1. The review_scores_rating is highly correlated with all the other fields except review_scores_location. Therefore, we could say that the location of the listing does not impact the review of the person who took the listing. Reviews were made on the basis of the a. Responsiveness of the host b. Ease in checking in c. Cleanliness of the listing and d. Accuracy of the description of the listings which are logical reasons.

2. A listing which was mentioned accurately in the description with proper photos and videos got more rating in their reviews.
